```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

pressures = np.array([1000,800,600])
temps = np.arange(250,305,5)
# print(pressures)
# print(temps)

def ln_theta(temp, pressure):
    """ln of Thompkins equation 36 for potential temperature
    inputs: temperature (K), pressure (Pa)
    outputs: ln(theta) -> array of floats"""
    
    Rd = 287.05 # J kg^-1 K^-1
    Cp = 1004 # J kg^-1 K^-1
    press0 = 1e5 # Pa
    return np.log(temp*(press0/pressure)**(Rd/Cp))

lntheta1 = ln_theta(temps, pressures[0])
lntheta2 = ln_theta(temps, pressures[1])
lntheta3 = ln_theta(temps, pressures[2])

fig,ax = plt.subplots(1,1)
ax.plot(temps,lntheta1, label="Pressure = 1000 hPa")
ax.plot(temps,lntheta2, label="Pressure = 800 hPa")
ax.plot(temps,lntheta3, label="Pressure = 600 hPa")
ax.set_xlabel("Temperature (K)")
ax.set_ylabel("lntheta")
ax.grid(True)
ax.legend()
```
